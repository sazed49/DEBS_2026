mkrum round 20

```python
# =============================== Defining global variables ========================#

DATASET_NAME = "MineSigns"   # must match what you added in sampling.py/distribute_dataset()
MODEL_NAME   = "CNNMineSigns"     # or "CNNMineSigns" if you added it

DD_TYPE = "NON_IID"               # start with IID for sanity
ALPHA = 1

NUM_PEERS = 10                 # IMPORTANT: start small for debugging
FRAC_PEERS = 1                # select all peers each round

SEED = 7
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

CRITERION = nn.CrossEntropyLoss()

GLOBAL_ROUNDS = 20
LOCAL_EPOCHS  = 5
TEST_BATCH_SIZE = 64
LOCAL_BS = 32
LOCAL_LR = 1e-3
LOCAL_MOMENTUM = 0.9

NUM_CLASSES = 13

# If you have sign names, put them here; otherwise this simple mapping works
LABELS_DICT = {str(i): i for i in range(NUM_CLASSES)}

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("DEVICE:", DEVICE)

# Label flip attack configuration (change as you like)
SOURCE_CLASS = 2
TARGET_CLASS = 0

# for EXTREME_NON_IID only (ignored in IID/NON_IID)
CLASS_PER_PEER = 3
SAMPLES_PER_CLASS = 200
RATE_UNBALANCE = 1
```

```python
RULE = 'mkrum'
ATTACK_TYPE='label_flipping'
MALICIOUS_BEHAVIOR_RATE = 1
for atr in [0.4]:
    run_exp(dataset_name = DATASET_NAME, model_name = MODEL_NAME, dd_type = DD_TYPE, num_peers = NUM_PEERS, 
            frac_peers = FRAC_PEERS, seed = SEED, test_batch_size = TEST_BATCH_SIZE,
                criterion = CRITERION, global_rounds = GLOBAL_ROUNDS, local_epochs = LOCAL_EPOCHS, local_bs = LOCAL_BS, 
                 local_lr = LOCAL_LR, local_momentum = LOCAL_MOMENTUM, labels_dict = LABELS_DICT, device = DEVICE,
                attackers_ratio = atr, attack_type=ATTACK_TYPE, 
                 malicious_behavior_rate = MALICIOUS_BEHAVIOR_RATE, rule = RULE,
                source_class = SOURCE_CLASS, target_class = TARGET_CLASS,
               class_per_peer = CLASS_PER_PEER, samples_per_class = SAMPLES_PER_CLASS, 
               rate_unbalance = RATE_UNBALANCE, alpha = ALPHA, resume = False)
```
